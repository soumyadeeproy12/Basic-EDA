```python
def grade_average(dataframe):
    return((dataframe['G1']+dataframe['G2']+dataframe['G3'])/3 )
#getting the average grade
```

```python
df['average_grade'] = df.apply(lambda row: grade_average(row), axis = 1)
```

```python
df['average_grade'].plot(kind = "density")
#density curve of average grade
```

```python

sns.barplot(x = "studytime", y = "average_grade", data = df)
#studytime vs average study time analysis
```

```python
sns.pairplot(df, hue="Walc")
#all sorts pair visualization plots possible
```

```python
sns.barplot(x = "sex", y = "average_grade", hue = "internet", data = df)
#avaerage grades for internet users
```

```python
sns.jointplot(x="age", y="average_grade", data=df, kind="kde");
#agewise grade plot
```

```python
df['alcoholic'] = df.apply(lambda row: alcoholic(row), axis = 1)
```

```python
sns.barplot(x = "sex", y = "average_grade", hue = "alcoholic", data = df)
#grade vs alcoholic students analysis
```

```python
sns.relplot(x="average_grade", y="freetime", hue="romantic", data=df);
#somehow trickiest analysis, it is for the students who have free time and romantically involved versus their grades
#there can be many more analysis established from the datasets
```
